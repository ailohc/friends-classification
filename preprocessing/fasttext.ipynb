```python
import os
import numpy as np
import re
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
DEFAULT_PATH='./../data/'
TRAIN_PATH = os.path.join(DEFAULT_PATH, 'friends_train.json')
TEST_PATH = os.path.join(DEFAULT_PATH, 'friends_test.json')
```

```python
with open(TRAIN_PATH) as train_file:
    train = json.load(train_file)
with open(TEST_PATH) as test_file:
    test = json.load(test_file)
```

```python
train = [lst for sub in train for lst in sub]
test = [lst for sub in test for lst in sub]
```

```python
train[0]
```

```python
with open(os.path.join(DEFAULT_PATH, 'cleaned_friends_train.txt'), "a") as f:
    for line in train:
        orig_utterance = line['utterance']
        orig_label = line['emotion']
        
        if type(orig_utterance) == str:
            cleaned_utterance = re.sub(u'’',u"'", orig_utterance)
            
            line = "__label__" + orig_label + " " + cleaned_utterance
            print(line)
            
            f.write(line + '\n')
```

```python
with open(os.path.join(DEFAULT_PATH, 'cleaned_friends_test.txt'), "a") as f:
    for line in test:
        orig_utterance = line['utterance']
        orig_label = line['emotion']
        
        if type(orig_utterance) == str:
            cleaned_utterance = re.sub(u'’',u"'", orig_utterance)
            
            line = "__label__" + orig_label + " " + cleaned_utterance
            print(line)
            
            f.write(line + '\n')
```
